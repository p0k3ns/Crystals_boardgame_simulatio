```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
#card deck
d = []
for i in ['Hearts','Spades','Clubs','Diamonds']:
    for j in ['2','3','4','5','6','7','8','9','10','Jack','Queen','King','Ace']:
        d.append(j +' of '+ i)
deck = pd.DataFrame(d)
```

```python
def shuffle(deck):
    shuffled_deck = deck.sample(frac=1)
    shuffled_deck = shuffled_deck.reset_index(drop=1)
    return shuffled_deck
```

```python
def first_table(shuffled_deck):
    table = shuffled_deck.iloc[0:5]
    shuffled_deck = shuffled_deck.iloc[5:]
    table = table.reset_index(drop=1)
    shuffled_deck = shuffled_deck.reset_index(drop=1)
    return table, shuffled_deck
```

```python
def choose_and_add(table,shuffled_deck):
    table = table.drop(table.index[np.random.randint(0,5)])
    table = table.append(shuffled_deck.iloc[0])
    shuffled_deck = shuffled_deck.iloc[1:]
    shuffled_deck = shuffled_deck.reset_index(drop=1)
    table = table.reset_index(drop=1)
    return table, shuffled_deck
```

```python
deck = pd.DataFrame(d)
shuffled_deck = shuffle(deck)
table, shuffled_deck = first_table(shuffled_deck)
table, shuffled_deck = choose_and_add(table, shuffled_deck)
```
